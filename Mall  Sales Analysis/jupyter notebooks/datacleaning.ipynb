```python
import pandas as pd
import re as re
```

```python
data=pd.read_csv(r"C:\Users\ronob\Desktop\retail sales data sets\mall_data.csv")
df=pd.DataFrame(data)
df
```

```python
#cleaning invoice number
df['invoice_no']=df['invoice_no'].str.strip()
#checking for concistency 
pattern = r'^I\d+$'  # Example pattern for an "I" prefix followed by digits
consistency_check = all(re.match(pattern, invoice_id) for invoice_id in df['invoice_no'])
print("Consistency Check:", consistency_check)
```

```python
#standardising variables 
df['shopping_mall'] = df['shopping_mall'].str.strip().str.title()
df['payment_method'] = df['payment_method'].str.strip().str.title()
df['category'] = df['category'].str.strip().str.title()
df['gender'] = df['gender'].str.strip().str.title()
```

```python
#checking for null values
null_rows = df[df.isnull().any(axis=1)]
print("Rows with Null Values:")
print(null_rows)
#verifying the absence of the null values
all_columns_null = df.isnull().any().any()

if all_columns_null:
    print("There are null values in the DataFrame.")
else:
    print("There are no null values in the DataFrame.")
```

```python
# checking for Age Outliers
age_summary = df['age'].describe()
print("Age Summary:", age_summary)
# Handling Inconsistent Categorical Values
df['category'] = df['category'].str.strip().str.title()
df['payment_method'] = df['payment_method'].str.strip().str.title()
df['shopping_mall'] = df['shopping_mall'].str.strip().str.title()
df['gender'] = df['gender'].str.strip().str.title()
# Handling Data Format Consistency
df['invoice_date'] = pd.to_datetime(df['invoice_date'], dayfirst=True)

print(df.head())
```

```python
#saving the cleaned data frame into a new csv
cleaned_filename = "Cleaned_Mall_Data.csv"
df.to_csv(cleaned_filename, index=False)  
print(f"Cleaned data saved to {cleaned_filename}")
```
